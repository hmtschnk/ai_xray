```python
#version_v5

STANDARD_OVERLAY_SIZE = (512, 512)
ORIGINAL_DISPLAY_IMG_KEY = "original_display_image"
PATHOLOGY_THRESHOLD = 0.7
ANATOMY_THRESHOLD = 0.7
WEIGHT = "densenet121-res224-all"
#MODEL_CLASS = baseline_models.chestx_det.PSPNet

import cv2
import dicom2jpg
import torchxrayvision as xrv
import skimage.io
import torch
import torchvision
import os
import json
import uuid
import shutil
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from safetensors.torch import save_file, load_file
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget
from skimage.transform import resize
import warnings
import datetime
from io import BytesIO
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
import matplotlib.patheffects as path_effects

# Suppress non-critical warnings from libraries
warnings.filterwarnings("ignore", category=UserWarning)


# --- Helper Functions and Classes ---

class ImagePreprocessor:
    """Handles all image loading and preprocessing for models and visualization."""
    @staticmethod
    def load_and_preprocess_for_display(img_path):
        """
        Loads, normalizes, and resizes an image for visualization.

        Parameters:
        - img_path (str): The absolute or relative path to the image file.

        Outputs:
        - np.ndarray: A NumPy array of the preprocessed image, shape (H, W, 3) and normalized to [0, 1].
        """
        img = skimage.io.imread(img_path)
        if len(img.shape) == 3 and img.shape[2] == 4:
            img = img[..., :3]
        
        img_normalized = img.astype(np.float32)
        img_normalized = (img_normalized - img_normalized.min()) / (img_normalized.max() - img_normalized.min() + 1e-8)
        
        return resize(img_normalized, STANDARD_OVERLAY_SIZE, anti_aliasing=True, preserve_range=True)

    @staticmethod
    def load_and_preprocess_for_model(img_path, target_size):
        """
        Loads, converts to grayscale, normalizes, and resizes for model input.

        Parameters:
        - img_path (str): The path to the image file.
        - target_size (int): The target dimension (H and W) for the model's input.

        Outputs:
        - np.ndarray: A NumPy array ready for model input, shape (1, H, W).
        """
        img = skimage.io.imread(img_path)
        if len(img.shape) == 3:
            img = img.mean(2)
        
        img = xrv.datasets.normalize(img, 255)
        img = resize(img, (target_size, target_size), anti_aliasing=True, preserve_range=True)
        
        return img[None, ...]

    def convert_dicom_to_jpg(dicom_path, output_dir, jpg_filename):
        """
        Converts a DICOM file to JPG and saves it with a specified name.

        Parameters:
        - dicom_path (str): The path to the input DICOM file.
        - output_dir (str): The directory for the converted JPG file.
        - jpg_filename (str): The desired filename for the output JPG.

        Outputs:
        - str: The path to the newly created JPG file.
        
        Error Handling:
        - Raises RuntimeError: If dicom2jpg conversion fails.
        - Raises FileNotFoundError: If the converted file is not found.
        """
        try:
            dicom2jpg.dicom2jpg(dicom_path, output_dir)
            converted_file_found = False
            for file in os.listdir(output_dir):
                if file.lower().endswith(".jpg"):
                    shutil.move(os.path.join(output_dir, file), os.path.join(output_dir, jpg_filename))
                    converted_file_found = True
                    break
            if not converted_file_found:
                raise FileNotFoundError("JPG conversion failed. No JPG file found.")
            return os.path.join(output_dir, jpg_filename)
        except Exception as e:
            raise RuntimeError(f"Error converting DICOM to JPG: {e}")

# --- Core Processor Class ---

class XRayProcessor:
    """Processes X-ray images to detect diseases and generate overlays."""
    
    def __init__(self, xray_file, output_path, unique_id):
        self.xray_file = xray_file
        self.output_path = output_path
        self.id = str(unique_id)
        
        now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        try:
            os.makedirs(self.output_path, exist_ok=True)
            print(f"[{now}][API] Output directory ensured\n----- {self.output_path}\n----- PASS\n")
        except Exception as e:
            print(f"[{now}][API] Output directory ensured\n----- {self.output_path}\n----- FAIL: {e}\n")
        
        self.jpg_path = os.path.join(output_path, f"{self.id}.jpg")
        
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        print(f"[{now}][API] Using device\n----- {self.device}\n----- PASS\n")
        
        self.cls_model = xrv.models.DenseNet(weights=WEIGHT).to(self.device).eval()
        self.seg_model = xrv.baseline_models.chestx_det.PSPNet().to(self.device).eval()

    def _get_image_for_processing(self):
        """Prepares the image file for analysis and overlay generation."""
        now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print(f"[{now}][API] Preparing image file\n----- {self.xray_file}")
        
        try:
            file_ext = os.path.splitext(self.xray_file)[-1].lower()
            if file_ext == ".dcm":
                jpg_path = convert_dicom_to_jpg(self.xray_file, self.output_path, os.path.basename(self.jpg_path))
                print(f"[{now}][API] Image prepared for ID\n----- {self.id}\n----- {jpg_path}\n----- PASS\n")
                return jpg_path
            elif file_ext in [".jpg", ".png", ".jpeg"]:
                shutil.copy2(self.xray_file, self.jpg_path)
                print(f"[{now}][API] Image prepared for ID\n----- {self.id}\n----- {self.jpg_path}\n----- PASS\n")
                return self.jpg_path
            else:
                raise ValueError("Unsupported file format. Only DICOM, JPG, and PNG are supported.")
        except Exception as e:
            print(f"[{now}][API] Image prepared\n----- FAIL: {e}\n")
            raise

    def _analyze_pathologies(self, img_path):
        """Runs the classification model to get disease predictions."""
        now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print(f"[{now}][API] Analyzing X-ray image for pathologies\n----- {img_path}")
        try:
            img = ImagePreprocessor.load_and_preprocess_for_model(img_path, target_size=224)
            input_tensor = torch.from_numpy(img).unsqueeze(0).float().to(self.device)
            with torch.no_grad():
                outputs = self.cls_model(input_tensor)
            results = dict(zip(self.cls_model.pathologies, map(float, outputs[0].detach().cpu().numpy())))
            print(f"[{now}][API] Analysis complete\n----- PASS\n")
            return results
        except Exception as e:
            print(f"[{now}][API] Analysis complete\n----- FAIL: {e}\n")
            raise

    def _generate_disease_overlays(self, img_path, original_image):
        """Generates Grad-CAM overlays with single bounding box per disease only."""
        import cv2
        from skimage.transform import resize
    
        # Prepare input tensor
        img = ImagePreprocessor.load_and_preprocess_for_model(img_path, target_size=224)
        input_tensor = torch.from_numpy(img).unsqueeze(0).float().to(self.device)
    
        target_layer = self.cls_model.features.denseblock3.denselayer16.conv2
        cam = GradCAM(model=self.cls_model, target_layers=[target_layer])
    
        # Ensure original image has 3 channels
        if len(original_image.shape) == 2:
            original_image_rgb = np.stack([original_image]*3, axis=-1)
        else:
            original_image_rgb = original_image.copy()
    
        overlays = {ORIGINAL_DISPLAY_IMG_KEY: torch.from_numpy(original_image_rgb).float()}
    
        # Parameters
        pathology_threshold = PATHOLOGY_THRESHOLD
        box_color = (255, 0, 0)   # BGR
        box_thickness = 1
        alpha = 0.3
    
        for i, pathology in enumerate(self.cls_model.pathologies):
            # Grad-CAM
            grayscale_cam = cam(input_tensor=input_tensor, targets=[ClassifierOutputTarget(i)])[0]
            norm_heatmap = (grayscale_cam - grayscale_cam.min()) / (grayscale_cam.max() - grayscale_cam.min() + 1e-8)
            cam_overlay_resized = resize(norm_heatmap, original_image_rgb.shape[:2], preserve_range=True)
            cam_overlay_rgb = np.stack([cam_overlay_resized]*3, axis=-1)
    
            # Blend heatmap
            boxed_img = (1-alpha)*original_image_rgb + alpha*cam_overlay_rgb
            boxed_img = np.clip(boxed_img, 0, 1)
    
            # Only draw bounding box for disease overlays
            mask = cam_overlay_resized >= pathology_threshold
            mask_uint8 = (mask.astype(np.uint8) * 255)
            contours, _ = cv2.findContours(mask_uint8, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            if len(contours) > 0:
                contour = max(contours, key=cv2.contourArea)
                x, y, w, h = cv2.boundingRect(contour)
                boxed_img_uint8 = (boxed_img * 255).astype(np.uint8)
                cv2.rectangle(boxed_img_uint8, (x, y), (x + w, y + h), box_color, box_thickness)
                overlays[pathology] = torch.from_numpy(boxed_img_uint8 / 255.0).float()
            else:
                overlays[pathology] = torch.from_numpy(boxed_img).float()
    
        return overlays





    def _generate_anatomical_overlays(self, img_path, original_image):
        """Generates segmentation overlays for anatomical regions."""
        img = ImagePreprocessor.load_and_preprocess_for_model(img_path, target_size=512)
        input_tensor = torch.from_numpy(img).unsqueeze(0).float().to(self.device)
        
        overlays = {ORIGINAL_DISPLAY_IMG_KEY: torch.from_numpy(original_image).float()}
        with torch.no_grad():
            pred = self.seg_model(input_tensor).cpu().squeeze(0).numpy()
            
        for i, region in enumerate(self.seg_model.targets):
            heatmap = pred[i]
            norm_heatmap = (heatmap - heatmap.min()) / (heatmap.max() - heatmap.min() + 1e-8)
            mask = norm_heatmap >= ANATOMY_THRESHOLD
            active_area = norm_heatmap.copy()
            active_area[~mask] = np.nan
            overlays[region] = torch.from_numpy(active_area).float()
            
        return overlays
        
    def process(self):
        """Main method to process the X-ray image."""
        try:
            img_path = self._get_image_for_processing()
            original_display_img = ImagePreprocessor.load_and_preprocess_for_display(img_path)
            
            # 1. Analyze for pathologies and save results
            results = self._analyze_pathologies(img_path)
            json_path = os.path.join(self.output_path, f"{self.id}.json")
            output_data = {"id": self.id, "input_img": os.path.basename(self.xray_file), "results": results}
            
            now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            try:
                with open(json_path, "w", encoding="utf-8") as f:
                    json.dump(output_data, f, indent=4)
                print(f"[{now}][API] Disease analysis results\n----- {json_path}\n----- PASS\n")
            except Exception as e:
                print(f"[{now}][API] Disease analysis results\n----- FAIL: {e}\n")
                raise

            # 2. Generate and save overlays
            print(f"[{now}][API] Generating overlays")
            
            try:
                disease_overlays = self._generate_disease_overlays(img_path, original_display_img)
                disease_path = os.path.join(self.output_path, f"DO-{self.id}.safetensors")
                save_file(disease_overlays, disease_path)
                print(f"[{now}][API] Disease Grad-CAM\n----- {disease_path}\n----- PASS\n")
            except Exception as e:
                print(f"[{now}][API] Disease Grad-CAM\n----- FAIL: {e}\n")
                raise
                
            try:
                anatomical_overlays = self._generate_anatomical_overlays(img_path, original_display_img)
                anatomy_path = os.path.join(self.output_path, f"AO-{self.id}.safetensors")
                save_file(anatomical_overlays, anatomy_path)
                print(f"[{now}][API] Anatomical segmentation overlays\n----- {anatomy_path}\n----- PASS\n")
            except Exception as e:
                print(f"[{now}][API] Anatomical segmentation overlays\n----- FAIL: {e}\n")
                raise
            
            print(f"[{now}][API] Processing complete for ID\n----- {self.id}\n----- PASS\n")
            return self.id, results
        except Exception as e:
            now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            print(f"[{now}][API] Processing complete for ID\n----- {self.id}\n----- FAIL: {e}\n")
            return None


# --- Core Visualizer Class ---

class XRayVisualizer:
    """Visualizes X-ray images with disease and anatomical overlays."""
    
    def __init__(self, uuid_str, output_path):
        self.uuid = uuid_str
        self.output_path = output_path
        self.disease_results = None
        self.disease_data = None
        self.anatomical_data = None
        self.base_image_display = None
        self._load_data()

    def _load_data(self):
        """Loads all necessary data from the specified output path."""
        now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print(f"[{now}][API] Loading data for visualization\n----- {self.uuid}\n")
        try:
            json_path = os.path.join(self.output_path, f"{self.uuid}.json")
            disease_path = os.path.join(self.output_path, f"DO-{self.uuid}.safetensors")
            anatomy_path = os.path.join(self.output_path, f"AO-{self.uuid}.safetensors")
            
            with open(json_path, "r", encoding="utf-8") as f:
                self.disease_results = json.load(f)["results"]
            
            self.disease_data = load_file(disease_path)
            self.anatomical_data = load_file(anatomy_path)
            
            self.base_image_display = self.disease_data[ORIGINAL_DISPLAY_IMG_KEY].cpu().numpy()
            if len(self.base_image_display.shape) == 2:
                self.base_image_display = np.stack([self.base_image_display] * 3, axis=-1)
            elif self.base_image_display.shape[-1] == 4:
                self.base_image_display = self.base_image_display[..., :3]
            
            print(f"[{now}][API] Data successfully loaded\n----- PASS\n")
        except FileNotFoundError as e:
            print(f"[{now}][API] Data loading failed\n----- FAIL: {e}\n")
            raise FileNotFoundError(f"Missing file for UUID {self.uuid}: {e}")
        except Exception as e:
            print(f"[{now}][API] Data loading failed\n----- FAIL: {e}\n")
            raise RuntimeError(f"An unexpected error occurred loading data: {e}")

    def get_disease_results(self):
        """Returns the disease prediction results."""
        return self.disease_results


    def show_overlays(self, keys, alphas=None, colormaps=None, return_bytes=False):
        """Visualizes selected overlays (disease or anatomy) on the base image."""
        now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        if self.base_image_display is None:
            print(f"[{now}][API] Visualization failed: Image data not loaded.")
            return None
    
        print(f"[{now}][API] Generating visualization...")
    
        composite_image = self.base_image_display.copy()
        alphas = alphas if alphas is not None else [0.5] * len(keys)
        colormaps = colormaps if colormaps is not None else [cm.jet] * len(keys)
    
        if len(keys) != len(alphas) or len(keys) != len(colormaps):
            print(f"[{now}][API] Visualization failed: Length mismatch.\n----- FAIL\n")
            return None
    
        fig, ax = plt.subplots(figsize=(10, 10))
        ax.imshow(composite_image)
        ax.axis('off')
    
        label_positions = []  # track placed label centers
    
        def is_too_close(x, y, existing_positions, min_dist=30):
            for (ex, ey) in existing_positions:
                dist = np.sqrt((x - ex) ** 2 + (y - ey) ** 2)
                if dist < min_dist:
                    return True
            return False
    
        for key, alpha, cmap in zip(keys, alphas, colormaps):
            overlay_tensor = None
            if key in self.disease_data:
                overlay_tensor = self.disease_data[key]
            elif key in self.anatomical_data:
                overlay_tensor = self.anatomical_data[key]
        
            if overlay_tensor is None:
                print(f"[{now}][API] Warning: Key '{key}' not found. Skipping.")
                continue
        
            overlay_data = overlay_tensor.cpu().numpy()
        
            # Disease overlays (already with box)
            if key in self.disease_data:
                heatmap_rgb = overlay_data
                valid_mask = np.any(heatmap_rgb != 0, axis=-1)
        
            # Anatomy overlays (colormap applied)
            else:
                heatmap_rgba = cmap(overlay_data)
                heatmap_rgb = heatmap_rgba[..., :3]
                valid_mask = (~np.isnan(overlay_data)) & (np.sum(heatmap_rgb, axis=-1) > 0)
        
            for c in range(3):
                composite_image[..., c][valid_mask] = (
                    (1 - alpha) * composite_image[..., c][valid_mask] +
                    alpha * heatmap_rgb[..., c][valid_mask]
                )
    
            # --- Place label at max intensity point ---
            if np.any(valid_mask):
                if overlay_data.ndim == 3 and overlay_data.shape[2] == 3:
                    intensity_map = np.mean(overlay_data, axis=2)
                else:
                    intensity_map = overlay_data
                max_idx = np.nanargmax(intensity_map)
                max_y, max_x = np.unravel_index(max_idx, intensity_map.shape)
    
                # Adjust label position if overlapping
                label_x, label_y = max_x, max_y
                offset_step = 20
                attempts = 0
                while is_too_close(label_x, label_y, label_positions):
                    label_x += offset_step
                    label_y += offset_step
                    attempts += 1
                    if attempts > 10:
                        break
                label_positions.append((label_x, label_y))
    
                label_text = f"{'Disease' if key in self.disease_data else 'Anatomy'}: {key}"
                txt = ax.text(
                    label_x, label_y, label_text,
                    color='white', fontsize=15, weight='bold',
                    ha='center', va='center'
                )
                txt.set_path_effects([
                    path_effects.Stroke(linewidth=2, foreground='black'),
                    path_effects.Normal()
                ])
    
        ax.imshow(np.clip(composite_image, 0, 1))
        fig.subplots_adjust(left=0, right=1, top=1, bottom=0)
    
        if return_bytes:
            from io import BytesIO
            img_io = BytesIO()
            canvas = FigureCanvas(fig)
            canvas.print_png(img_io)
            plt.close(fig)
            img_io.seek(0)
            print(f"[{now}][API] Visualization rendered to bytes\n----- PASS\n")
            return img_io
        else:
            plt.show()
            print(f"[{now}][API] Visualization complete\n----- PASS\n")
            return None
```

```python
INFILE = r"00000022_001.jpg"
OUTFILE = r"xray_output"

# Predefined list of colormaps for overlays in visualization
PREDEFINED_COLORMAPS = [
    cm.jet, cm.viridis, cm.plasma, cm.inferno, cm.magma,
    cm.cividis, cm.cool, cm.hot, cm.spring, cm.summer
]

selected_diseases = ["Lung Opacity", "Pneumonia", "Left Scapula","Right Scapula", "Heart","Right Lung"]
```

```python
# --- Example Usage (Cell 1: Processing) ---

if __name__ == '__main__':
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    try:

        processor = XRayProcessor(INFILE, OUTFILE, 'test123')
        processed_uuid, results = processor.process()

        if processed_uuid:
            print(f"\n[{now}] Processing complete. The ID '{processed_uuid}' is ready for visualization.")
            print(f"Copy the ID and run the code in the next cell to visualize the results.")
            print(f"\n[{now}] Overall process status\n----- PASS\n")
        else:
            print(f"[{now}] Overall process status\n----- FAIL: X-ray processing failed.\n")

    except (FileNotFoundError, ValueError, RuntimeError) as e:
        print(f"\n[{now}] An error occurred: {e}\n----- FAIL\n")
    except Exception as e:
        print(f"\n[{now}] An unexpected error occurred: {e}\n----- FAIL\n")
```

```python
from IPython.display import display
from PIL import Image
import io

try:
    # Define transparency and colormaps for each overlay key
    alphas = [0.5] * len(selected_diseases)
    colormaps = [PREDEFINED_COLORMAPS[i % len(PREDEFINED_COLORMAPS)] for i in range(len(selected_diseases))]

    # Create visualizer instance
    visualizer = XRayVisualizer(processed_uuid, OUTFILE)
    
    # Generate the overlay image bytes for the selected diseases
    image_io = visualizer.show_overlays(
        keys=selected_diseases,
        alphas=alphas,
        colormaps=colormaps,
        return_bytes=True  # ensure the method returns image bytes
    )

    if image_io:
        # Load image from bytes
        image = Image.open(image_io)
        # Display image inline in Jupyter notebook
        display(image)

except Exception as e:
    print(f"[ERROR] render_overlay failed: {e}")
```
